**I.** Imagine o setor elétrico brasileiro simplificado e formado pelos seguintes conjuntos:

- `USINAS` = conjunto de todas as usinas cadastradas.
- `RENOVAVEIS` = usinas que são renováveis.
- `DISPONIVEIS_HOJE` = usinas que estão operando hoje.

Para futuras decisões sobre planejamento energético, considere:

```{code-block} python3
USINAS = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1", "Termica_Gas", "Termica_Carvao"}
RENOVAVEIS = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1"}
DISPONIVEIS_HOJE = {"UHE_Sobradinho", "EOL_Ventos", "FV_Solar1", "Termica_Gas"}
```

Determine: 

a. quais usinas são renováveis e estão disponíveis hoje;

b. quais usinas estão paradas;

c. se há alguma usina disponível hoje e que não seja renovável.

````python
# Criando os conjuntos das três usinas

usinas = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1", "Termica_Gas", "Termica_Carvao"}
renovaveis = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1"}
disponiveis_hoje = {"UHE_Sobradinho", "EOL_Ventos", "FV_Solar1", "Termica_Gas"}
````

````python
# Imprimindo as usinas

print(f"USINAS = {usinas}")
print(f"RENOVAVEIS = {renovaveis}")
print(f"DISPONIVEIS_HOJE = {disponiveis_hoje}")
````

````python
### Questão 1 - item 'a':
## Determine quais usinas são renováveis e estão disponíveis hoje.
# Para saber quais usinas estão nos conjuntos renovaveis e disponiveis_hoje, é necessário determinar a interseção entre ambos os conjuntos mencionados.

renovaveis_disponiveis_hoje = set()

for usina in disponiveis_hoje:
    if usina in renovaveis:
        renovaveis_disponiveis_hoje.add(usina)
print(f"Renováveis disponíveis hoje = {renovaveis_disponiveis_hoje}")
````

````python
### Questão 1 - item 'b':
## Determine quais usinas estão paradas.
# Para saber quais usinas estão paradas, é necessário verificar quais usinas não estão na interseção entre os conjuntos usinas e disponiveis_hoje.

paradas = set()

for usina in usinas:
    if usina not in disponiveis_hoje:
        paradas.add(usina)

print(f"Usinas paradas = {paradas}")
````

````python
### Questão 1 - item 'c':
## Determine se há alguma usina disponível hoje e que não seja renovável.
# Para saber se há alguma usina não renovável disponível, é necessário obter a diferença entre os conjuntos disponiveis_hoje e renovaveis.

nao_renovaveis = set()

for usina in disponiveis_hoje:
    if usina not in renovaveis:
        nao_renovaveis.add(usina)

#nao_renovaveis = disponiveis_hoje.difference(renovaveis)
print(f"Usinas não renováveis = {nao_renovaveis}")
````

**II.** Supondo os seguintes conjuntos de fontes por região:

```{code-block} python3
fontes_NE = {"eolica", "solar", "hidreletrica"}
fontes_SE = {"hidreletrica", "termica_gas", "solar"}
```

Determine o conjunto:

a. das fontes comuns no Nordeste e Sudeste;

b. todas as fontes dessas regiões;

c. todas as fontes típicas do Nordeste;

````python
# Criando os conjuntos das duas fontes

fontes_NE = {"eolica","solar","hidreletrica"}
fontes_SE = {"hidreletrica","termica_gas","solar"}
````

````python
# Imprimindo as fontes

print(f"fontes_NE = {fontes_NE}")
print(f"fontes_SE = {fontes_SE}")
````

````python
### Questão 2 - item 'a':
## Determine o conjunto das fontes comuns no Nordeste e Sudeste.
# Para saber quais usinas estão em ambos os conjuntos, é necessário obter a interseção entre eles.

fontes_comuns = set()

for fonte in fontes_NE:
    if fonte in fontes_SE:
        fontes_comuns.add(fonte)
        
print(f"Fontes mútuas = {fontes_comuns}")
````

````python
### Questão 2 - item 'b':
## Determine o conjunto de todas as fontes dessas regiões.
# Para saber todas as fontes dessas regiões, é necessário obter a união entre os conjuntos fontes_NE e fontes_SE

fontes = set()

for conjunto in(fontes_NE, fontes_SE):
    for fonte in conjunto:
        fontes.add(fonte)
        
print(f"Fontes = {fontes}")
````

````python
### Questão 2 - item 'c':
## Determine o conjunto de todas as fontes típicas do Nordeste.
# Para saber todas as fontes específicas da região Nordeste, é necessário obter a diferença entre os conjuntos fontes_SE e fontes_NE.

tipicas_NE = set()

for fonte in fontes_NE:
    if fonte not in fontes_SE:
        tipicas_NE.add(fonte)

print(f"Fontes exclusivas do Nordeste = {tipicas_NE}")
````

**III** Considere os seguintes grupos de pacientes de um hospital:

- `PACIENTES` = todos atendidos no mês.
- `DIABETES` = diagnosticados com diabetes.
- `HIPERTENSOS` = diagnosticados com hipertensão.
- `ACOMP_FAMILIA` = estão em acompanhamento.

Defina:

```{code-block} python3
PACIENTES = {"Ana", "Bruno", "Carlos", "Dora", "Erica", "Fábio", "Gil", "Helena"}
DIABETES = {"Ana", "Carlos", "Erica"}
HIPERTENSOS = {"Carlos", "Dora", "Fabio"}
ACOMP_FAMILIA = {"Ana", "Dora", "Helena", "Gil"}
```

Determine: 

a. pacientes com as duas condições;

b. pacientes crônicos (pelo menos uma condição);

c. pacientes crônicos (que estão em acompanhamento);

````python
PACIENTES = {"Ana","Bruno","Carlos","Dora","Erica","Fábio","Gil","Helena"}
DIABETES = {"Ana","Carlos","Erica"}
HIPERTENSOS = {"Carlos","Dora","Fábio"}
ACOMP_FAMILIA = {"Ana","Dora","Helena","Gil"}
````

````python
print(f"PACIENTES = {PACIENTES}")
print(f"DIABETES = {DIABETES}")
print(f"HIPERTENSOS = {HIPERTENSOS}")
print(f"ACOMP_FAMILIA = {ACOMP_FAMILIA}")
````

````python
### Questão 3 - 'item a'
## Determine os pacientes com as duas condições.
# Para saber os pacientes que têm ambas as condições, é necessário obter a interseção entre os conjuntos DIABETES e HIPERTENSOS.

nabeiradocaixao = set()

for paciente in PACIENTES:
    if paciente in DIABETES and paciente in HIPERTENSOS:
        nabeiradocaixao.add(paciente)
        
print(f"Pacientes com um pé e 4 dedos na cova = {nabeiradocaixao}")
````

````python
### Questão 3 - 'item b'
## Determine os pacientes com pelo menos uma condição.
# Para determinar um paciente crônico, é necessário obter o todos os pacientes que estão em pelo menos um dos dois conjuntos.

cronicos = set()

for paciente in PACIENTES:
    if paciente in DIABETES or paciente in HIPERTENSOS:
        cronicos.add(paciente)
        
print(f"Pacientes Crônicos = {cronicos}")
````

````python
### Questão 3 - 'item c'
## Determine os pacientes crônicos em acompanhamento.
# Para determinar um paciente crônico em acompanhamento, é necessário obter todos os pacientes que estão em pelo menos um dos dois conjuntos, e que também estão no conjunto ACOMP_FAMILIA.

acompanhados = set()

for pacientes in (DIABETES, HIPERTENSOS):
    for paciente in pacientes:
        if paciente in ACOMP_FAMILIA:
            acompanhados.add(paciente)
            
print(f"Pacientes em acompanhamento = {acompanhados}")
````
