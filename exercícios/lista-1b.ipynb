**I.** Imagine o setor elétrico brasileiro simplificado e formado pelos seguintes conjuntos:

- `USINAS` = conjunto de todas as usinas cadastradas.
- `RENOVAVEIS` = usinas que são renováveis.
- `DISPONIVEIS_HOJE` = usinas que estão operando hoje.

Para futuras decisões sobre planejamento energético, considere:

```{code-block} python3
USINAS = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1", "Termica_Gas", "Termica_Carvao"}
RENOVAVEIS = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1"}
DISPONIVEIS_HOJE = {"UHE_Sobradinho", "EOL_Ventos", "FV_Solar1", "Termica_Gas"}
```

Determine: 

a. quais usinas são renováveis e estão disponíveis hoje;

b. quais usinas estão paradas;

c. se há alguma usina disponível hoje e que não seja renovável.

````python
# Criando os conjuntos das três usinas

usinas = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1", "Termica_Gas", "Termica_Carvao"}
renovaveis = {"UHE_BoaEsperanca", "UHE_Sobradinho", "PCH_Rio1", "EOL_Ventos", "FV_Solar1"}
disponiveis_hoje = {"UHE_Sobradinho", "EOL_Ventos", "FV_Solar1", "Termica_Gas"}
````

````python
# Imprimindo as usinas

print(f"USINAS = {usinas}")
print(f"RENOVAVEIS = {renovaveis}")
print(f"DISPONIVEIS_HOJE = {disponiveis_hoje}")
````

````python
### Questão 1 - item 'a':
## Determine quais usinas são renováveis e estão disponíveis hoje.
# Para saber quais usinas estão nos conjuntos renovaveis e disponiveis_hoje, é necessário determinar a interseção entre ambos os conjuntos mencionados.

renovaveis_disponiveis_hoje = renovaveis.intersection(disponiveis_hoje)
print(f"RENOVAVEIS_DISPONIVEIS_HOJE = {renovaveis_disponiveis_hoje}")
````

````python
### Questão 1 - item 'b':
## Determine quais usinas estão paradas.
# Para saber quais usinas estão paradas, é necessário verificar quais usinas não estão na interseção entre os conjuntos usinas e disponiveis_hoje.

paradas = usinas.difference(disponiveis_hoje)
print(f"PARADAS = {paradas}")
````

````python
### Questão 1 - item 'c':
## Determine se há alguma usina disponível hoje e que não seja renovável.
# Para saber se há alguma usina não renovável disponível, é necessário obter a diferença entre os conjuntos disponiveis_hoje e renovaveis.
nao_renovaveis = disponiveis_hoje.difference(renovaveis)
print(f"NAO_RENOVAVEIS = {nao_renovaveis}")
````
